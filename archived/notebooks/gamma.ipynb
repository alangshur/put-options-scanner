```python
import csv
import pandas as pd
import numpy as np
```

```python
target = 'scan/gamma-test-1.csv'
```

```python
# open file
f = open('../' + target, 'r')
data = np.array(list(csv.reader(f)))
f.close()
```

```python
# convert to dataframe
df = pd.DataFrame(data[:, 1:]).astype(np.float64)
df.index = data[:, 0]
df.columns = [
    'underlying', 'cost', 'width', 'put_be', 'call_be', 
    'relative_be_dist', 'net_delta', 'gamma_theta_ratio', 'std',
    'put_iv', 'call_iv', 'atm_iv', 'iv_skew', 'ret', 'ret_r2', 'corr',
    'hv', 'hvp_ivp_diff'
]
```

```python
# filter dataframe
df = df[df['hvp_ivp_diff'] > 0]
```
