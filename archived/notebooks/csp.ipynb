```python
import csv
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
target = 'scan/csp-equities-1.csv'
```

```python
# open file
f = open('../' + target, 'r')
data = np.array(list(csv.reader(f)))
f.close()
```

```python
# convert to dataframe
df = pd.DataFrame(data[:, 1:]).astype(np.float64)
df.index = data[:, 0]
df.columns = [
    'width', 'premium', 'max_loss', 'break_even', 'rr_ratio', 
    'prob_max_loss', 'prob_max_profit', 'ra_profit', 'delta', 
    'theta', 'vega', 'gamma'
]
```

```python
df_top = df.nlargest(25, 'rr_ratio')
df_top
```
