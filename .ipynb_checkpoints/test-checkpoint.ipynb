```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
f = open('data/SPY-1d.csv', 'r')
data = np.array(list(csv.reader(f)))
cl = data[:, 4].astype(np.float64) 
f.close()
```

```python
step_size = 5
hist_size = cl.shape[0]
reg_period = 60

steps = 0
last = None
reg_values = []
for i in range(reg_period, hist_size):
    
    if steps % step_size == 0:
        
        # compute linear regression
        x = np.arange(reg_period).reshape(-1, 1)
        y = cl[i - reg_period:i]
        reg = LinearRegression().fit(x, y)
        
        # compute error
        y_hat = reg.coef_ * np.arange(reg_period) + reg.intercept_
        error = np.linalg.norm(y - y_hat)
        last = (reg.coef_[0], error)
        reg_values.append(last)
    
    else:
        reg_values.append(last)
    
    steps += 1
```
